```python
import pandas as pd
```

```python
con="D:\\mypython\\trial\\Potable Water Drought Risk.xlsx"
Hazard = pd.read_excel(con,sheet_name='Hazard',header=1)
Hazard = pd.read_excel(con,sheet_name='Hazard',header=1)
Exposure=pd.read_excel(con,sheet_name='Exposure',header=1)
Susceptibility=pd.read_excel(con,sheet_name='Susceptibility',header=1)
Copingcapacity=pd.read_excel(con,sheet_name='Coping capacity',header=1)
Adaptive_capacity=pd.read_excel(con,sheet_name='Adaptive capacity',header=1)
```

```python
Hazard['Occurrence of warm'] = (Hazard['Occurrence of warm days (TX90p)']*0.1).tolist()
Hazard['No. of drought events']=(Hazard['No. of drought events - 1980 to 2019 (1980 -1999 , 25%, 2000 -2019, 75%)']*0.5).tolist()
Hazard['SPI_Intensity']=(Hazard['SPI (Intensity)']*0.4).tolist()
Hazard['Hazard_index']=(Hazard['Occurrence of warm']+Hazard['No. of drought events']+Hazard['SPI_Intensity']).tolist()
#Hazard.iloc[:,[0,1,2,9]]
```

```python
Exposure['Total_Population2'] = (Exposure['Total Population ']*0.25).tolist()
Exposure['Population_density2']=(Exposure['Population density']*0.3).tolist()
Exposure['drought affected People2']=(Exposure['Max # of drought affected People during 1980 to 2019 (1980 -1999, 25%, 2000 -2019, 75%)']*0.2).tolist()
Exposure['Environmental sensitive areas in DS ']=(Exposure['Environmental sensitive areas in DS ']*0.25).tolist()
Exposure['Total_DS area2'] = (Exposure['Total DS area']*0.1).tolist()
Exposure['Exposure_index']=(Exposure['Total_Population2']+Exposure['Population_density2']+Exposure['drought affected People2']+Exposure['Environmental sensitive areas in DS ']+Exposure['Total_DS area2']).tolist()
#Exposure.head()
#Hazard.iloc[:,[0,1,2,9]]
```

```python
Susceptibility['% Population fed  by water schemes (major, Rural, Community'] = (Susceptibility['% Population fed  by water schemes (major, Rural, Community)']*0.25).tolist()
Susceptibility['% of Water surface area']=(Susceptibility['% of Water surface area']*0.1).tolist()
Susceptibility['% of forest cover']=(Susceptibility['% of forest cover']*0.15).tolist()
Susceptibility['Susceptibility_Index']=(Susceptibility['% Population fed  by water schemes (major, Rural, Community']+Susceptibility['% of Water surface area']+Susceptibility['% of forest cover']).tolist()
#Susceptibility.head()
```

```python
Copingcapacity['% of unemployment'] = (Copingcapacity['% of unemployment']*0.2).tolist()
Copingcapacity['Poverty head count Index']=(Copingcapacity['Poverty head count Index']*0.3).tolist()
Copingcapacity['Dependency ratio']=(Copingcapacity['Dependency ratio']*0.25).tolist()
Copingcapacity['Adult education (% below ordinary level (O/L)']=(Copingcapacity['Adult education (% below ordinary level (O/L)']*0.15).tolist()
Copingcapacity['% of Female headed Household'] = (Copingcapacity['% of Female headed Household  ']*0.1).tolist()
Copingcapacity['Coping capacity_Index']=(Copingcapacity['% of unemployment']+Copingcapacity['Poverty head count Index']+Copingcapacity['Dependency ratio']+Copingcapacity['Adult education (% below ordinary level (O/L)']+Copingcapacity['% of Female headed Household']).tolist()
#Copingcapacity.head()
```

```python
Adaptive_capacity['Projects'] = (Adaptive_capacity['Projects']*1.0).tolist()
Adaptive_capacity['Adaptive_capacity_Index']=Adaptive_capacity['Projects']
#Adaptive_capacity.head()
#Adaptive_capacity.iloc[:,5]
```

```python
Summary=Hazard.iloc[:,[0,1,2,9]]
Summary
Exposure.head()
#x= Exposure.iloc[:,[12]]
Summary=Summary.assign(Exposure_index=Exposure.iloc[:,[12]],Susceptibility_Index=Susceptibility.iloc[:,[9]],
                       Coping_capacity_Index=Copingcapacity.iloc[:,9],Adaptive_capacity_Index=Adaptive_capacity.iloc[:,5])
#Summary
```

```python
Summary=Summary.assign(Exposure=Summary.loc[:,'Hazard_index']*Summary.loc[:,'Exposure_index'],
                       Vulnerability=(Summary['Susceptibility_Index']*.33)+(Summary['Coping_capacity_Index']*.33)+(Summary['Adaptive_capacity_Index']*.33))

Summary=Summary.assign(Risk=Summary['Exposure']*Summary['Vulnerability'])
#Summary['Exposure']=Summary.loc[:,'Hazard_index']*Summary.loc[:,'Exposure_index']
#Summary['Vulnerability']=(Summary['Susceptibility_Index']*.33)+(Summary['Coping_capacity_Index']*.33)+(Summary['Adaptive_capacity_Index']*.33)
#Summary['Risk']=Summary['Exposure']*Summary['Vulnerability']
#Summary
RiskMap=Summary.iloc[:,[0,2,10]]
RiskMap.sort_values(by=['Risk'],ascending=False)
```
